```python
#Importing files to be used in data visualization
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Loading the first five rows of the medical examination csv
df = pd.read_csv('medical_examination.csv')
df.head()
```

```python
# Creating a new column overweight
# Assign all values of overweight > 25 as 1 and < 25 as 0
# df['overweight'] = (df.weight / (df.height/100)**2).apply(lambda x:1 if x > 25 else 0)
df['overweight'] = (df['weight']/(df['height']/100)**2).apply(lambda x: 1 if x > 25 else 0)
df
```

```python
# Normalize the data by making 0 always good and 1 always bad.
# If the value of cholesterol or gluc is 1, make the value 0.
# If the value is more than 1, make the value 1.
df['cholesterol'] = df.cholesterol.apply(lambda x: 0 if x == 1 else 1)
df['gluc'] = df.gluc.apply(lambda x: 0 if x == 1 else 1)
df
```

```python
#Creating the data as a long format using the pandas melt function
df_cat = pd.melt(df, 
                 id_vars = 'cardio', 
                 var_name = 'variable', 
                 value_vars = ['active','alco','cholesterol', 'gluc','overweight','smoke'])

df_cat
```

```python
#Creating a seaborn catplot figure and assigning it variable fig
fig = sns.catplot(data=df_cat, 
                  kind="count",  
                  x="variable",
                  hue="value", 
                  col="cardio").set_axis_labels("variable", "total")
fig
```

```python
#Cleaning and filtering data based on the instructions
df_heat = df[(df['ap_lo']<=df['ap_hi']) &
    (df['height'] >= df['height'].quantile(0.025))&
    (df['height'] <= df['height'].quantile(0.975))&
    (df['weight'] >= df['weight'].quantile(0.025))&
    (df['weight'] <= df['weight'].quantile(0.975))
    ]
df_heat
```

```python
# Creating a correlation matrix of the df_heat data
corr = df_heat.corr()
corr
```

```python
# Generate a mask for the upper triangle
mask = np.triu(corr)
mask
```

```python
# setup a matplotlib figure
fig, ax = plt.subplots(figsize=(7, 5))

# Draw the heatmap with 'sns.heatmap()'
sns.heatmap(corr,
            mask=mask, 
            fmt='.1f',
            vmax=.3, 
            linewidths=.5,
            square=True, 
            cbar_kws = {'shrink':0.5},
            annot=True, 
            center=0)
```
